```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

arr1=[1,2,3,4,5,6,7]
arr2=[4,2,1,3,6,7,5]
#lineport needs a datafrmae to run,so we need panadas to make a dataframe.in the data frame we made a dictionary so identify the x and y variable
fr=pd.DataFrame({"arr1":arr1,"arr2":arr2})
#we are using line plot to show the line plot graph.we are passing the x and y value for the arr1 and arr2.most important,the 'data'.'data' mainly contains the infos of dataframe.
sns.lineplot(x="arr1",y="arr2",data=fr)
plt.title("just a random line plot graph")
plt.xlabel("x side")
plt.ylabel("y side")
plt.show()
```
