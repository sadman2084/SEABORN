```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#head() is used to show the amount wise data entered into head.
# load_data=sns.load_dataset("attention").head(20)
load_data=sns.load_dataset("attention")
#'subject' and 'score' is from attentio.csv file. and hue wil add multiple graphs
#size will  help to manage the graph size ,style will change the graph dotted.palette is the color to change
sns.lineplot(x="subject",y="score",data=load_data,hue="solutions",
             size=50,style="solutions",palette='Accent',
             markers=["o",">"],dashes=False,legend="brief")
#legend is there is a map which shows the lines,if it is false,the legend will disappear.It can be "False","Auto","full","brief"
plt.grid()
plt.title("about python")
plt.xlabel("subject")
plt.ylabel("score")
plt.show()

```
