Count Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data_load=sns.load_dataset("tips")

#data_load

sns.countplot(x="sex",data=data_load,hue="smoker",color="red",saturation=0.4)
#by setting up the value in x means the names will be in x axis and counts will be in y axis
#it counts the value of selected in x or y.this is the difference with count and bar plot
#to make it vertical,replace the x variable into y
#


plt.show()
```
