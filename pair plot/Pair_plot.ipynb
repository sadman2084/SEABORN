pair plot


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


data_load=sns.load_dataset("tips").head(20)

sns.pairplot(data_load,hue="sex")
plt.show()
```

```python
#if you want to show the specific column plots
sns.pairplot(data_load,vars=["total_bill","tip"],
             hue="sex",hue_order=["Female","Male"]
             ,palette="Accent")
plt.show()
#"vars" is a built-in variable
#"hue_order" shows "female" catagory first and then "male"
#
```

```python
#to set x parameter
sns.pairplot(data_load,x_vars=["total_bill","tip"])
plt.show()

```

```python
#to set y parameter
sns.pairplot(data_load,y_vars=["total_bill","tip"])
plt.show()
```

```python
#to make a density line
sns.pairplot(data_load,kind="reg",diag_kind="kde")
#"kind" can be scatter,kde,hist,reg
#"diag_kind" make a new plot on the previous plot and it can be used as scatter,kde,hist,reg.
plt.show()
```

```python
#to change the marker
sns.pairplot(data_load,markers=["*","<"])
plt.show()
```
