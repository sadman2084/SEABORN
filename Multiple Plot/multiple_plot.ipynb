Multiple Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

data_load=sns.load_dataset("tips")

fg=sns.FacetGrid(data_load,col="day",hue="sex")#loads the to fg

fg.map(plt.scatter,"total_bill","tip").add_legend()#then fg makes the graph into scatter,this graph can be shown with may other graphs

plt.show()
```

```python
fg=sns.FacetGrid(data_load,col="day",hue="sex")#loads the to fg

fg.map(plt.bar,"total_bill","tip",edgecolor="red").add_legend()#then fg makes the graph into scatter,this graph can be shown with may other graphs

plt.show()
```
