```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


data_load=sns.load_dataset("tips")

data_load

sns.violinplot(x="day",y="total_bill",
               data=data_load,hue="time",linewidth=1,
               color="red",palette="Accent"
               saturation=0.4)
#"linewidth" customizes the width of line
#color and palatte changes the color
#"saturation" uses to light the color
#

plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",
               data=data_load,order=["Dinner","Lunch"],color="red")
```

```python
sns.violinplot(x="day",y="total_bill",
               data=data_load,hue="sex",palette="Accent",split=True
               ,scale="width",inner="quartile")
#we can use "scale" as "count","width","area"
#"inner" can be use as "box","quartile","point","stick",none

```

```python
sns.violinplot(y="day",x="total_bill",
               data=data_load,hue="sex",color="blue",split=True
               ,scale="width",saturation=0.5)
```
