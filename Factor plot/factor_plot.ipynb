Factor PLot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data_load=sns.load_dataset("tips")

sns.catplot(x="size",y="tip",data=data_load,hue="sex",
            
            palette="Accent",height=10)

plt.show()
```

```python
sns.catplot(x="size",y="tip",data=data_load,hue="sex",
            
            palette="Accent",height=10,kind="bar")


plt.show()
```

```python
sns.catplot(x="size",y="tip",data=data_load,hue="sex",
            
            palette="Accent",height=10,kind="point")


plt.show()
```

```python
sns.catplot(x="size",y="tip",data=data_load,hue="sex",
            
            palette="Accent",height=10,kind="boxen")


plt.show()
```
