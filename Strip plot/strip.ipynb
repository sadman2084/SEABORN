```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


data_load=sns.load_dataset("tips")



sns.stripplot(x="day",y="total_bill",data=data_load,color="red",
              hue="sex",palette="Accent",linewidth=1
              ,edgecolor="red",jitter=1,size=5,marker="<",alpha=0.9)
#"edgecolor" changes the color of border of the dots
#"jitter" increses the distance between dots
#"size" increses the size of the dot
#"alpha" uses to light the color
#

plt.show()
```

to show single data(horizontal)

```python
sns.stripplot(x=data_load["total_bill"])
```

to show single data(vertial)

```python
sns.stripplot(y=data_load["total_bill"])
```
