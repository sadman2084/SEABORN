Histogram plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

load_data=sns.load_dataset("car_crashes")
#
# load_data

#sns.displot(load_data["alcohol"])
sns.displot(x=load_data.alcohol,data=load_data,kde=True,
            rug=True,color="green",log_scale=True)

#"kde" makes a curve line on the bars
#"rug" creates a density line bottom of the bar
#"color" changes the color of bars
#"log_scale" shows the value in the logarithm form
plt.show()
```

```python
sns.displot(x=load_data.alcohol,y=load_data.speeding,
            data=load_data,bins=[2,4,6,8,10,12],color="pink")
#"bins" helps in the x axis values to show,for understanding,2 4 6 8 is in the x axis
plt.show()
```
